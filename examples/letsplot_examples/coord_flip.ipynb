```kotlin
%useLatestDescriptors
%use lets-plot
%use dataframe
%use ggdsl(0.1.4-dev-29)
```

```kotlin
LetsPlot.getInfo() // This prevents Krangl from loading an obsolete version of Lets-Plot classes.
```

```kotlin
var mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-kotlin/master/docs/examples/data/mpg.csv")
mpg.head(3)
```

```kotlin
val p = letsPlot(mpg.toMap()) + theme(axisTitleX="blank")
```

```kotlin
val boxPlot = p + 
    geomBoxplot(color="#579673", fill="#9AC0B3", size=1.5) {
        x = asDiscrete("class", order=1, orderBy="..middle..")
        y = "hwy" 
    } + 
    ggsize(750, 300)
    
boxPlot + ggtitle("Default")    
```

```kotlin
mpg.create { plot {
    boxplot(`class`, hwy) {
        orderBy(Stat.MIDDLE)
        borderLineColor(Color.fromHex("#579673"))
        fillColor(Color.fromHex("#9AC0B3"))
        borderLineWidth(1.5)
    }
    layout {
        title = "Default"
        size = 750 to 300
        theme { 
            xAxis.title {
              blank = true
            }
        }
    }  
} }
```

```kotlin
boxPlot + ggtitle("Flipped") + coordFlip()
```

```kotlin
mpg.create { plot {
    boxplot(`class`, hwy) {
        orderBy(Stat.MIDDLE)
        borderLineColor(Color.fromHex("#579673"))
        fillColor(Color.fromHex("#9AC0B3"))
        borderLineWidth(1.5)
    }
    coordFlip()
    layout {
        title = "Flipped"
        size = 750 to 300
        theme { 
            xAxis.title {
              blank = true
            }
        }
    }  
} }
```

```kotlin
val barPlot = p + 
    geomBar(size=1.5, alpha=0.7, width=0.7) {
        x = asDiscrete("manufacturer", orderBy="..count..")
        fill = "class"
        color = "class"
    } +
    ggsize(800, 300)
    
barPlot + ggtitle("Default")
```

```kotlin
mpg.groupBy {
    manufacturer and `class`
}.aggregate { 
    count() into "count"
}.join(mpg.groupBy {
    manufacturer
}.aggregate { 
    count() into "countManufacturer"
}, "manufacturer").sortByDesc("countManufacturer")
.create { plot {
   bar { 
       // todo orderBy
       x(manufacturer)
       y("count"<Int>())
       alpha(.7)

       width(.7)
       color(`class`)
       borderLine{
           width(1.5)
           color(`class`)
       }
       position = Position.Stack
   }
   layout{
       size = 800 to 300
       title = "Default"
   } 
} }
```

```kotlin
barPlot + ggtitle("Flipped") + coordFlip()
```

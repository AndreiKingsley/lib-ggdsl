# Plot bunch

*plotBunch()* allows to show a collection of plots on one figure. Each plot in the collection can have arbitrary location and size. There is no automatic layout inside the bunch.

```kotlin
%useLatestDescriptors
%use lets-plot

@file:DependsOn("org.apache.commons:commons-math3:3.6.1")
%use ggdsl(0.2.4-dev-5)
```

```kotlin
import kotlin.random.Random
import org.apache.commons.math3.distribution.MultivariateNormalDistribution
```

```kotlin
val cov : Array<DoubleArray> = arrayOf(doubleArrayOf(1.0, 0.0),
                                        doubleArrayOf(0.0, 1.0))
val means : DoubleArray = doubleArrayOf(0.0, 0.0)
val xy = MultivariateNormalDistribution(means, cov).sample(400)

val xs = xy.map { it[0] }
val ys = xy.map { it[1] }
```

### View this data as a scatter plot and as a histogram

```kotlin
val myScatter = plot {
    point {
        x(xs)
        y(ys)
        color(Color.BLACK)
        alpha(.4)
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
        yAxisLabel = "y"
    }
}
myScatter
```

```kotlin
val myHistogram = plot {
    histogram(xs) {
        color(Color.named("dark_magenta"))
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
    }
}
myHistogram
```

### Combine both plots in one figure

```kotlin
val scaleX = continuousPos(-3.5 to 3.5)
plotBunch { 
    add(plot {
    histogram(xs) {
        x(scaleX)
        color(Color.named("dark_magenta"))
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
    }
}, 0, 0) 
    add(plot {
    point {
        x(xs.scaled(scaleX))
        y(ys)
        color(Color.BLACK)
        alpha(.4)
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
        yAxisLabel = "y"
    }
}, 0, 200)
}
```

### Adjust visuals of the bunch figure

```kotlin
val upperTheme = theme {
    blankAxes()
    axis.text { }
    yAxis.title { }
    panel.grid.majorYLine {
        blank = true
    }
}
val lowerTheme = theme {
    blankAxes()
    yAxis.text { }
    axis.title { }
}
```

```kotlin
plotBunch { 
    add(plot {
    histogram(xs) {
        x(scaleX)
        color(Color.named("dark_magenta"))
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
        theme(upperTheme)
    }
}, 0, 0)
    add(plot {
    point {
        x(xs.scaled(scaleX))
        y(ys)
        color(Color.BLACK)
        alpha(.4)
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
        yAxisLabel = "y"
        theme(lowerTheme)
    }
}, 0, 200) 
}
```

### Adjust plot sizes

*addPlot()* method has two more (optional) parameters: *width* and *height*.

This values will override plot size earlier defined via *ggsize()* function.

```kotlin
plotBunch { 
    add(plot {
    histogram(xs) {
        x(scaleX)
        color(Color.named("dark_magenta"))
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
        theme(upperTheme)
    }
}, 0, 0, 600, 100)
    add(plot {
    point {
        x(xs.scaled(scaleX))
        y(ys)
        color(Color.BLACK)
        alpha(.4)
    }
    layout{
        size = 600 to 200
        xAxisLabel = "x"
        yAxisLabel = "y"
        theme(lowerTheme)
    }
}, 0, 100, 600, 300)
}
```

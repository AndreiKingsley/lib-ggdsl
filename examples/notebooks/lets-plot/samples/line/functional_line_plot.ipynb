```kotlin
val xs = (-2000..2000).map { it.toDouble() / 500.0 }
val function = { x: Double -> sin(x) * cos(x * 2 + 1) * sin(3 * x + 2.0) }
val ys = xs.map(function)

plot {
    line {
        x(xs)
        y(ys) {
            scale = continuous(-1.0..1.0)
        }
        hLine {
            yIntercept.constant(0.0)
            color = Color.RED
            type = LineType.DASHED
        }
    }
}
```

```kotlin
%useLatestDescriptors
%use kandy
%use dataframe
```

```kotlin
val category = (List(50) { "easy" } + List(50) { "medium" } + List(30) { "hard" }).shuffled()
val halfSize = category.size / 2
val year = List(halfSize) { "2022" } + List(halfSize) { "2023" }

val df = dataFrameOf(
    "category" to category,
    "year" to year
)

df.groupBy("category").plot { 
    countPlot("year")
}
```

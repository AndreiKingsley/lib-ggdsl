```kotlin
val random = java.util.Random(42)

val sample = List(1000) {random.nextGaussian()}

plot {
    histogram(sample) {
        alpha = 0.9
        fillColor = Color.WHITE
        y(Stat.density)
    }
    densityPlot(sample) {
        alpha = 0.6
        fillColor = Color.hex(0xFF6666)
    }
}
```

```kotlin
val xs = (-1000..1000).map { it / 250.0f }
val function = { x: Float -> sin(2 * x) * cos(x / 2 - 3.0) }
val yActual = xs.map(function)
val yLow = yActual.map { it - 0.25 }
val yHigh = yActual.map { it + 0.25 }
```

```kotlin
plot {
    ribbon { 
        x(xs)
        yMin(yLow)
        yMax(yHigh)
        fillColor = Color.LIGHT_BLUE
    }
}
```

```kotlin
val random = java.util.Random(42)

val sample = List(1000) {random.nextGaussian()}

plot { 
    statBin(sample, binsOption = BinsOption.byNumber(15)) {
        bars {
            alpha = 0.9
            x(Stat.x)
            y(Stat.count)
        }
        line {
            x(Stat.x)
            y(Stat.count)
            color = Color.RED
            width = 1.5
            type = LineType.LONGDASH
        }
    }
}
```

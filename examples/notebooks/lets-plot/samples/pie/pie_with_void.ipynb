# Pie Chart With "Void" Style

```kotlin
%useLatestDescriptors
%use dataframe, kandy
```

```kotlin
val df = dataFrameOf(
    "platform" to listOf("Linux", "MacOS", "Windows"), 
    "count" to listOf(30, 239, 566)
)
```

```kotlin
df.plot { 
    pie { 
        slice(count)
        fillColor(platform)
    }
    layout {
        style(Style.Void)
    }
}
```

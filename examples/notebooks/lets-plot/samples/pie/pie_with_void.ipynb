# Pie Chart With "Void" Style

```kotlin
%use dataframe, kandy(0.7.0-dev-51, 0.3.0-dev-26)
```

```kotlin
val df = dataFrameOf(
    "target" to listOf("Linux", "MacOS", "Windows"), 
    "count" to listOf(30, 239, 566)
)
```

```kotlin
df.plot { 
    pie { 
        slice("count")
        fillColor("target")
    }
    layout {
        style(Style.Void)
    }
}
```

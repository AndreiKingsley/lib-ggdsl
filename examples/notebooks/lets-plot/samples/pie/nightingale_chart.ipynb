# Nightingale Chart

```kotlin
%useLatestDescriptors
%use dataframe, kandy
```

```kotlin
val month = listOf("Jan", "Feb", "Mar", "May", "Apr")
val amount = listOf(34.4, 25.1, 33.6, 20.0, 15.9)
```

```kotlin
plot { 
    pie { 
        slice(amount)
        fillColor(month, "month")
        size(amount) {
            scale = continuous(10.0..25.0)
            legend.type = LegendType.None
        }
    }
    layout {
        style(Style.Void)
    }
}
```

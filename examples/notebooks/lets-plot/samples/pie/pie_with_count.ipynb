# Pie Chart With Count

```kotlin
%use dataframe, kandy(0.7.0-dev-51, 0.3.0-dev-26)
```

```kotlin
val df = dataFrameOf(
    "continent" to listOf(
        "EU", "AF", "SA", "OC", "EU", "AF", "SA", "AF", "AS", "SA",
        "OC", "OC", "SA", "NA", "AF", "NA", "EU", "AF", "OC", "SA",
        "AF", "SA", "OC", "EU", "AF"
    )
)
```

```kotlin
df.plot {
    statCount(continent) {
        pie {
            slice(Stat.count)
            fillColor(Stat.x named "continent")
            size = 25.0
        }
    }
    layout {
        style(Style.Void)
    }
}
```

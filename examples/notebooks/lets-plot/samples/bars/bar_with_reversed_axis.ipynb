```kotlin
%useLatestDescriptors
%use kandy
```

```kotlin
val dataset = dataFrameOf(
    "task" to listOf("a", "b", "c", "d", "e"),
    "time" to listOf(30, 25, 20, 35, 28)
)
```

```kotlin
dataset.plot {
    bars {
        x(task)
        y(time) {
            scale = continuous(transform = Transformation.REVERSE)
        }
        fillColor = Color.hex("#07C3F2")
    }
}
```

```kotlin
%use kandy(0.5.0-rc-1)
%use dataframe(0.12.0)
@file:Repository("https://packages.jetbrains.team/maven/p/kds/kotlin-ds-maven")
@file:DependsOn("org.jetbrains.kotlinx:kotlin-statistics-jvm:0.0.2")
```

```kotlin
// Create a random density plot and save it in a variable "p".
val rand = java.util.Random(123)
val n = 400
val data = mapOf (
    "rating" to List(n/2) { rand.nextGaussian() } + List(n/2) { rand.nextGaussian() * 1.5 + 1.5 },
    "cond" to List(n/2) { "A" } + List(n/2) { "B" }
)
```

```kotlin
val myP = plot(data) {
    groupBy("cond") {
         densityPlot("rating", trim = true) {
             fillColor = Color.GREY
             alpha = 0.6
             borderLine.color("cond")
         }
    }
    layout.size = 500 to 250
}

myP
```

```kotlin
import java.io.File
```

### Export PNG to file

```kotlin
val pathPNG = myP.save("myDensity.png")
```

```kotlin
javax.imageio.ImageIO.read(File(pathPNG))
```

### Export SVG to file

```kotlin
val pathSVG = myP.save("myDensity.svg")
```

```kotlin
HTML(File(pathSVG).readText())
```

### Export HTML to file

```kotlin
val pathHTML = myP.save("myDensity.html")
```

```kotlin
HTML(File(pathHTML).readText())
```

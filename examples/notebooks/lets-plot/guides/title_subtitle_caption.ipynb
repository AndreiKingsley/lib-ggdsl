```kotlin
%use kandy
%use dataframe
```

```kotlin
val mpgDf =
    DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-kotlin/master/docs/examples/data/mpg.csv")
mpgDf.head()
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout.title = "The plot title using 'layout.title'"
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title using 'Layout context'"
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title"
        subtitle = "The plot subtitle"
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title"
        subtitle = "The plot subtitle"
        caption = "The plot caption"
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title"
        subtitle = "The plot subtitle"
        caption = "The plot caption"
        theme {
            global.title {
                color = Color.BLUE
            }
        }
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title"
        subtitle = "The plot subtitle"
        caption = "The plot caption"
        theme {
            global {
                text {
                    fontFace = FontFace.ITALIC
                }
                title {
                    fontFace = FontFace.ITALIC
                }
            }
            plotCanvas {
                title {
                    color = Color.BLUE
                }
                subtitle {
                    fontFace = FontFace.ITALIC
                }
                caption {
                    fontFace = FontFace.ITALIC
                }
            }
        }
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title"
        subtitle = "The plot subtitle"
        caption = "The plot caption"
        theme {
            plotCanvas {
                title {
                    color = Color.BLUE
                }
                subtitle {
                    color = Color.RED
                }
                caption {
                    color = Color.named("dark_green")
                }
            }
        }
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(drv)
    }
    layout {
        title = "The plot title:\nFuel efficiency for most popular models of car"
        subtitle = "The plot subtitle:\nPoints are colored by the type of drive train"
        caption = "The plot caption:\nmpg dataset"
        theme {
            plotCanvas {
                subtitle {
                    color = Color.GREY
                }
            }
        }
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(cty) { legend.name = "City mileage" }
        symbol(drv) { legend.name = "Drive type" }
        size = 4.0
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(cty) { legend.name = "City mileage" }
        symbol(drv) { legend.name = "Drive type" }
        size = 4.0
    }
    layout.theme {
        legend.position = LegendPosition.Bottom
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(cty) { legend.name = "City mileage\n(mpg)" }
        symbol(drv) { legend.name = "Drive type\n(front/4/rear wheel)" }
        size = 4.0
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(cty) { legend.name = "City mileage\n(mpg)" }
        symbol(drv) { legend.name = "Drive type\n(front/4/rear wheel)" }
        size = 4.0
    }
    layout.theme {
        legend.position = LegendPosition.Bottom
    }
}
```

```kotlin
mpgDf.plot {
    points {
        x(displ)
        y(hwy)
        color(cty) { legend.name = "City mileage\n(mpg)" }
        symbol(drv) {
            legend {
                type = LegendType.DiscreteLegend(nRow = 3)
                name = "Drive type\n(front/4/rear wheel)"
            }
        }
        size = 4.0
    }
    layout.theme {
        legend.position = LegendPosition.Bottom
    }
}
```

## Customize points

```kotlin
val type = listOf("a", "b", "c", "d")
val df = dataFrameOf(
    "bs" to List(100) { random.nextInt(21) + 9 },
    "sp" to List(100) { random.nextInt(215) + 95 },
    "tp" to List(100) { type.random() },
    "sz" to List(100) { random.nextInt(30) + 5}
)
```

```kotlin
df.plot { 
    x(bs) { scale = continuous(limits = 5..35) }
    y(sp)

    points { 
        symbol = Symbol.DIAMOND
        color(tp)
        border { 
            color = Color.rgb(0, 0, 0)
        }

         size(sz) {
            scale = continuous(domain = 5..35, range = 10.0..35.0)
        }
    }


    layout {
        title.text = "Title"
        size = 800 to 600
    }
}
```

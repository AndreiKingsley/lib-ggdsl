```kotlin
%useLatestDescriptors
%use ggdsl-echarts
%use ggdsl-dataframe
```

## Basic Line Chart

```kotlin
val df = dataFrameOf(
        "month" to listOf("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"),
        "temp" to listOf(-2, -1, 1, 5, 14, 22, 24, 24, 15, 6, 3, 0)
    )
```

```kotlin
df.create {  
    plot {
        x(month)
        line {
            y(temp)
        }
    }
}
```

## Smothed Line

```kotlin
df.create { 
    plot {
        x(month)
        y(temp)
        line {
            smooth(true)
        }
    }
}
```

## Line with symbols

```kotlin
df.create { 
    plot { 
        x(month)
        y(temp)
        line {
            symbol(Symbol.CIRCLE)
        }
    }
}
```

```kotlin
df.create { 
    plot { 
        x(month)
        y(temp)
        line {
            symbol(Symbol.rect(size = 20.0, rotate = 45))
        }
    }
}
```

## Customize line

```kotlin
df.create { 
    plot {
        x(month)
        line {
            y(temp)
            color(temp.scaled(continuous(temp.min() to temp.max())))
            width(3.5)
            alpha(0.9)
            lineType(LineType.SOLID)
            shadowBlur(2)
        }
    }
}
```

## Step line

```kotlin
val df = dataFrameOf(
        "month" to listOf("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"),
        "start" to listOf(-9, -9, -4, 1, 7, 11, 15, 13, 8, 3, -2, -6),
        "middle" to listOf(-2, -1, 1, 5, 9, 12, 15, 14, 11, 6, 3, 0),
        "end" to listOf(4, 4, 6, 9, 13, 17, 19, 19, 16, 12, 8, 5)
    )
```

```kotlin
df.create { 
    plot {
        x(month)
        line {
            y(start)
            step(Step.START)
        }
        line {
            y(middle)
            step(Step.MIDDLE)
        }
        line {
            y(end)
            step(Step.END)
        }
    }
}
```

## Stacked line

```kotlin
val df = dataFrameOf(
    "year" to listOf("2015", "2016", "2017", "2018", "2019", "2020", "2021"),
    "apple" to listOf(233, 215, 229, 265, 260, 274, 365),
    "google" to listOf(74, 89, 110, 136, 160, 181, 256),
    "microsoft" to listOf(93, 85, 89, 110, 125, 143, 168)
    )
```

```kotlin
df.create { 
    plot {
        layout { 
            title.text = "Revenue comparison"
            tooltip { trigger = Trigger.AXIS }
        }
        x(year)
        val comp = stack("company")
        line {
            name("Apple revenue")
            y(apple)
            stack = comp
        }
        line { 
            name("Google revenue")
            y(google)
            stack = comp
        }
        line { 
            name("Microsoft revenue")
            y(microsoft)
            stack = comp
        }
    }
}
```
